```julia
y=0:(2pi/1000+.000001):2pi |> x->(sin.(x*8));
x=y+(0:(2pi/1000+.000001):2pi |> x->(sin.(x*40)));

responsetype = Lowpass(10; fs=1000)
designmethod = FIRWindow(hanning(64))
res=filt(digitalfilter(responsetype, designmethod), x);

#plot(0:1000,x[1:1001])
#plot(x, label="noisy")
plot(res, label="recover");
plot(y,label="primary");
legend(bbox_to_anchor=(-.1,1.1),loc=0)
```

```julia
y=0:(2pi/1000+.000001):2pi |> x->(sin.(x*3));
x=y+(0:(2pi/1000+.000001):2pi |> x->(sin.(x*10)));

responsetype = Bandpass(1, 4; fs=1000)
designmethod = Butterworth(4)
res=filt(digitalfilter(responsetype, designmethod), x)
plot(res, label="recover");
plot(y,label="primary");
legend(bbox_to_anchor=(-.1,1.1),loc=0)
```
